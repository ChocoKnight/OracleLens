```python
import pandas as pd 
import numpy as np
import requests
import csv
import time
```

```python
matches2025 = pd.read_csv("FilteredMatchTeamsDateWinner2025.csv")
matches2025.head()
```

```python
len(matches2025)
```

```python
matches2025.columns.tolist()
```

```python
tp = 0
tn = 0
fp = 0
fn = 0
correct = 0
current = 0

for i, row in matches2025.iterrows():
    # print(row)
    # print(f"Processing match {i+1} of {len(matches2025)}: {row['id']} {row['team_one']} vs {row['team_two']} on {row['date']}")
    matches = "http://localhost:8000/predict"
    predictionParams = {'teamOne': row['team_one'], 'teamTwo': row['team_two'], 'date': row['date']}
    prediction = requests.get(matches, params=predictionParams)
    prediction_data = prediction.json()

    # print(row)
    # print(prediction_data)
    
    if prediction_data['blueWin'] == row['blueWin']:
        correct += 1

    if prediction_data['blueWin'] == 1 and row['blueWin'] == 1:
        tp += 1
        print(f"True Positive: {row['team_one']} vs {row['team_two']} on {row['date']}")
    elif prediction_data['blueWin'] == 1 and row['blueWin'] == 0:
        fp += 1
        print(f"False Positive: {row['team_one']} vs {row['team_two']} on {row['date']}")
    elif prediction_data['blueWin'] == 0 and row['blueWin'] == 0:
        tn += 1
        print(f"True Negative: {row['team_one']} vs {row['team_two']} on {row['date']}")
    elif prediction_data['blueWin'] == 0 and row['blueWin'] == 1:
        fn += 1 
        print(f"False Negative: {row['team_one']} vs {row['team_two']} on {row['date']}")

    current += 1

    # print accuracy precision recall f1 
    accuracy = correct / current
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    print(f"Accuracy: {accuracy:.2f} Precision: {precision:.2f} Recall: {recall:.2f} F1 Score: {f1_score:.2f}")
    time.sleep(0.1)
```

```python
# accuracy precision recall f1 
accuracy = correct / len(matches2025) if len(matches2025) > 0 else 0
precision = tp / (tp + fp) if (tp + fp) > 0 else 0
recall = tp / (tp + fn) if (tp + fn) > 0 else 0
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

print(f"Total Matches: {len(matches2025)}")
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")   
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1_score:.2f}")

# with 4 layer lstm 
# Total Matches: 788
# Accuracy: 0.45
# Precision: 0.58
# Recall: 0.44
# F1 Score: 0.50

# with 5 layer lstm 
# Total Matches: 788
# Accuracy: 0.56
# Precision: 0.62
# Recall: 0.79
# F1 Score: 0.69
```
